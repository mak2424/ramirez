```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

import term_aver_plot
import theta
```

```python
#print(os.getcwd()) #выводит текущую папку
# filename = input("Enter filename: ")
filename = "merged_result.txt"
try:
	open("data/" + filename, "r")
	print("File opened: \"" + filename + "\"!")
	input_file = Path("data/" + filename)
except IOError:
	print("Error opening file \"" + filename + "\"!")

```

```python
# Настройка шрифта для всех элементов графика
plt.rcParams.update({
	"font.size": 16,            # размер шрифта
	"font.family": "serif",      # шрифт, можно "Times New Roman", "DejaVu Serif"
	"axes.labelsize": 18,        # размер подписей осей
	"axes.titlesize": 20,        # размер заголовка
	"xtick.labelsize": 14,       # размер подписей делений X
	"ytick.labelsize": 14        # размер подписей делений Y
})
```

```python
df = pd.read_csv(input_file, sep=r"\s+", engine="python")

x = df.iloc[:, 6]   # фрустрации
y = df.iloc[:, 5]   # вырождение


# --- Построение графиков ---

# plt.plot(x, y, color="blue", label=f"h/j = {H_ext}")
plt.plot(x, y, 'k+')
# plt.legend(loc="best")

# Настройка графика
plt.xlabel("frustrations")
#plt.ylabel(df.columns[4])
# plt.ylabel(r"1/$\chi$")
plt.ylabel(r"$g$")
# plt.ylabel(r"$\chi/N$")
# plt.title("(c)")
# plt.title(name_without_extension)

plt.yscale("log")


# === ДИАПАЗОНЫ ===
# plt.xlim(0, )     # X от 0 до 10
# plt.ylim(0, )    # Y от -5 до 20

plt.tight_layout()

# Сохранение графика в файл
# plt.savefig(f"{folder_path}/Xi_{name_without_extension}.png", dpi=300, bbox_inches='tight')  # формат можно изменить: .png, .jpg, .pdf, .svg
plt.savefig(f"g(f).png", dpi=300, bbox_inches='tight')  # формат можно изменить: .png, .jpg, .pdf, .svg
plt.show()

plt.close()
```
