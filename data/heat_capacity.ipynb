```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Настройка шрифта для всех элементов графика
plt.rcParams.update({
	"font.size": 16,            # размер шрифта
	"font.family": "serif",      # шрифт, можно "Times New Roman", "DejaVu Serif"
	"axes.labelsize": 18,        # размер подписей осей
	"axes.titlesize": 20,        # размер заголовка
	"xtick.labelsize": 14,       # размер подписей делений X
	"ytick.labelsize": 14        # размер подписей делений Y
})
```

```python
# --- Загрузка данных ---
folder = "data/term/ferro/"
folder_path = Path(folder)   # или "."
files = list(folder_path.glob("term_gem_glass64_J16_9*.dat"))
filename = str(files[0])
name_without_extension = files[0].stem[5:]
open(filename, "r")
input_file = Path(filename)


# теплоемкость/T
# --- имя без расширения ---
name_without_extension = input_file.stem[5:]
df = pd.read_csv(input_file, sep=r"\s+", engine="python")

x = df.iloc[:, 0]   # первый столбец — X (температура)
y = df.iloc[:, 3]   # четверый столбец (теплоемкость) / T

power = 3
koef = 2
y2 = (x**power) / koef

mask = (x >= 0.1) & (x <= 0.6)

# Подбор полинома
degree = 4
coefficients = np.polyfit(x[mask], y[mask], degree)
poly = np.poly1d(coefficients)

# Генерация точек для гладкого графика
x_smooth = np.linspace(x[mask].min(), x[mask].max(), 100)
y_smooth = poly(x_smooth)
plt.plot(x_smooth, y_smooth, color="red", label=f'Полином {degree}-й степени')

# --- Построение графиков ---
plt.plot(x[mask], y[mask], color="blue", label=f"C")
plt.plot(x[mask], y2[mask], color="black", label=f"(T^{power}) / {koef}")
plt.legend(loc="best")

# Настройка графика
plt.xlabel("T")
plt.ylabel(r"$C$")
plt.title(name_without_extension)

plt.tight_layout()
plt.show()
plt.close()
```

```python
import glob
import os

# Путь к папке с файлами
folder_path = 'data/term/ferro/'

# Поиск всех файлов
files = glob.glob(os.path.join(folder_path, 'term*.dat'))

plt.figure(figsize=(10, 6))

mask = (x >= 0.15) & (x <= 0.6)

for file in files:
    df = pd.read_csv(file, sep=r"\s+", engine="python")
    # Извлечение имени файла без расширения для легенды
    label = os.path.basename(file).replace('term_gem_glass64_', '').replace('_H0.0', '').replace('.dat', '')
    Cmax = df.iloc[:, 3].max()
    Tc = df.iloc[:, 0][df.iloc[:, 3] == Cmax].iloc[0]
    df_normalized = df.copy()
    df_normalized = df_normalized.iloc[:, 0]/Tc
    # plt.plot(df_normalized[mask], df.iloc[:, 3][mask], label=label, linewidth=2)
    plt.plot(df_normalized, df.iloc[:, 3], label=label, linewidth=2)

plt.xlim(0.1, 0.4)      # 0.4
plt.ylim(0, 0.11)

plt.xlabel('T')
plt.ylabel('C')
plt.title('ferro')


# Ваш исходный код с добавлением сортировки по конкретному X
target_x = 0.4  # Задайте нужное значение X

# Функция для получения Y при заданном X
def get_y_at_x(line, x_target):
    xdata = line.get_xdata()
    ydata = line.get_ydata()
    idx = np.argmin(np.abs(xdata - x_target))
    return ydata[idx]

# Получаем текущие линии и их данные
handles, labels = plt.gca().get_legend_handles_labels()

# Сортируем по значению Y при target_x
pairs = list(zip(handles, labels))
sorted_pairs = sorted(pairs, 
                     key=lambda pair: get_y_at_x(pair[0], target_x), 
                     reverse=True)  # reverse=True для убывания

handles_sorted, labels_sorted = zip(*sorted_pairs)

# Применяем отсортированную легенду
plt.legend(handles_sorted, labels_sorted, bbox_to_anchor=(1.05, 1), loc='upper left')

# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
